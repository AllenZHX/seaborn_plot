```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="darkgrid")
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
bins = np.linspace(0,60,13)
g.map(plt.hist, "total_bill",bins=bins)
```

```python
g = sns.FacetGrid(tips, col="sex", hue="time", margin_titles=True)
g.map(plt.scatter, "total_bill", "tip").add_legend()
```

```python
g = sns.FacetGrid(tips, col="day",aspect=.8)
g.map(sns.boxplot, "time","total_bill")
```
