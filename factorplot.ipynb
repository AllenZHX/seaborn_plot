```python
%matplotlib inline
import seaborn as sns
```

```python
sns.set(style="ticks")
exercise = sns.load_dataset("exercise")
g = sns.factorplot(x="time", y="pulse", hue="kind", data=exercise)
```

```python
g = sns.factorplot(x="time", y="pulse", hue="kind", data=exercise, col="diet")
```

```python
titanic = sns.load_dataset("titanic")
g = sns.factorplot("alive", col="deck", col_order=["A","B","C","D","E","F","G"],col_wrap=4, data = titanic[titanic.deck.notnull()], kind="count")
```

```python
g = sns.factorplot(x="age",y="embark_town", hue="sex",row="class",row_order=["First","Second","Third"],data=titanic[titanic.embark_town.notnull()], \
                  orient="h", aspect=3, palette="Set1",kind="violin",split=True)
```

```python
g = sns.factorplot(x="who",y="survived", col="class",col_order=["First","Second","Third"], data=titanic, kind="bar")
(g.set_axis_labels("","Survival Rate").set_xticklabels(["Men","Women","Children"]).set(ylim=(0, 1)).despine(left=True))
```
