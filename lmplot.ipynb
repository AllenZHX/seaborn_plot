```python
%matplotlib inline
import seaborn as sns
```

```python
sns.set(style="ticks")
```

```python
df = sns.load_dataset("anscombe")
```

```python
'''data looks like:
dataset       x 	y
I       	10.0 	8.04
I       	8.0 	6.95
II      	13.0 	7.58
II      	9.0 	8.81
III     	11.0 	8.33
III     	14.0 	9.96
'''
```

```python
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data = df,
          col_wrap=4, ci=100, palette="muted", size=4,
          scatter_kws={"s":50, "alpha":1})
```

```python
sns.lmplot(x="x", y="y", hue="dataset", data = df,
          ci=100, palette="Set1",)
```
