```python
%matplotlib inline
import seaborn as sns
```

```python
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.stripplot(x = tips["total_bill"])
```

```python
ax = sns.stripplot(x = tips["day"],y = tips["total_bill"], order=["Thur","Fri","Sat","Sun"])
#ax = sns.striplot(x="day",y="total_bill", data=tips)
```

```python
ax = sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, order=["Thur","Fri","Sat","Sun"])
```

```python
ax = sns.stripplot(x="day", y="total_bill", hue="sex", data=tips, jitter=True, order=["Thur","Fri","Sat","Sun"])
```

```python
ax = sns.stripplot(x="day", y="total_bill", hue="sex", data=tips, jitter=True, \
                   order=["Thur","Fri","Sat","Sun"], split=True)
```
