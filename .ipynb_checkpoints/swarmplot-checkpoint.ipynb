```python
%matplotlib inline
import seaborn as sns
```

```python
sns.set_style("whitegrid")
tips = sns.load_dataset("tips")
ax = sns.swarmplot(x=tips["total_bill"])
```

```python
ax = sns.swarmplot(x="day", y="total_bill", data=tips, order=["Thur","Fri","Sat","Sun"])
```

```python
ax = sns.swarmplot(x="day", y="total_bill", hue="sex",data=tips, order=["Thur","Fri","Sat","Sun"])
```

```python
ax = sns.swarmplot(x="day", y="total_bill",hue="sex",data=tips, \
                   order=["Thur","Fri","Sat","Sun"], split=True)
```

```python
ax = sns.violinplot(x="day",y="total_bill", data=tips)
ax = sns.swarmplot(x="day",y="total_bill", data=tips,color="white")
```
