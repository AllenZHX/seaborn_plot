```python
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(0,2*np.pi,31)
data = np.sin(x)
ax = sns.tsplot(data=data)
```

```python
data01 = np.sin(x) + np.random.rand(10,31) + np.random.randn(10,1)
ax = sns.tsplot(data=data01, err_style="ci_bars", interpolate=False)
```

```python
gammas = sns.load_dataset("gammas")
ax = sns.tsplot(time="timepoint", value="BOLD signal", unit="subject", condition="ROI", data=gammas)
```
