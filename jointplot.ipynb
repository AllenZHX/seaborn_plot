```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="white", color_codes=True)
tips = sns.load_dataset("tips")
g = sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
g = sns.jointplot(x="total_bill",y="tip", data=tips, kind="reg")
```

```python
g = sns.jointplot(x="total_bill",y="tip", data=tips, kind="hex")
```

```python
iris = sns.load_dataset("iris")
g = sns.jointplot("sepal_width","petal_length", data=iris, kind="kde",space = 0,color="r")
```

```python
from scipy.stats import spearmanr
g = sns.jointplot(x="size",y="total_bill", data=tips, stat_func=spearmanr)
```

```python
g = (sns.jointplot("sepal_length", "sepal_width",data=iris, color="k").plot_joint(sns.kdeplot, n_levels=6))
```
