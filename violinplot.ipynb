```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
tips = sns.load_dataset("tips")
```

```python
ax = sns.violinplot(x=tips["total_bill"])
```

```python
ax = sns.violinplot(x="day", y="tip", hue="sex", data=tips, palette="Set2", bw=.2, \
                    split=True, scale="count",inner="stick", order=["Thur","Fri","Sat","Sun"])
```
